```python
import requests
import re
```

```python
r = requests.get("https://api.nusmods.com/v2/2022-2023/moduleList.json")
```

```python
r_data = r.json()
```

```python
r_data
```

```python
all_modules = [r.get("moduleCode") for r in r_data]
all_modules
```

```python
# for regular CS mods
data = {}

cs_mods = list(filter(lambda x: re.match("CS\d{4}", x), all_modules))

for mod in cs_mods:
    data[mod] = requests.get("https://api.nusmods.com/v2/2022-2023/modules/" + mod + ".json").json()
```

```python
# for other requirements
data = {}

other_mods = ["IS1108", "ES2660", "MA2001", "MA1521", "ST2334"]

for mod in other_mods:
    data[mod] = requests.get("https://api.nusmods.com/v2/2022-2023/modules/" + mod + ".json").json()
```

```python
ge_mods = list(filter(lambda x: x.startswith("GE"), all_modules))
ge_mods
```

```python
# for other requirements
data = {}

for mod in ge_mods:
    data[mod] = requests.get("https://api.nusmods.com/v2/2022-2023/modules/" + mod + ".json").json()
```

```python
import json

with open("ge_mods.json", "w") as f:
    json.dump(data, f)
```
